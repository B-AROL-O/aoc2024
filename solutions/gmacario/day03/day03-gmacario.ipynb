```python
# Part 1: Test data

import re
p = re.compile(r"mul\([0-9]+\,[0-9]+\)")

test_line = "xmul(2,4)%&mul[3,7]!@^do_not_mul(5,5)+mul(32,64]then(mul(11,8)mul(8,5))"

total = 0
for m in p.finditer(test_line):
    # print(f"DEBUG: m = {m}")
    # print(m.start(), m.group())
    operands = re.findall("[0-9]+", m.group())
    # print(f"DEBUG: operands={operands}")
    total += int(operands[0]) * int(operands[1])

print(f"Part 1: TEST: TOTAL = {total}")
```

```python
FILENAME = "day03-input-gmacario.txt"

total = 0
with open(FILENAME) as f:
    for line in f:
        # print(line)
        for m in p.finditer(line):
            # print(m.start(), m.group())
            operands = re.findall("[0-9]+", m.group())
            # print(f"DEBUG: operands={operands}")
            total += int(operands[0]) * int(operands[1])

print(f"Part 1: TOTAL = {total}")
```

```python
# Part 2: Test data

def part2(line2):
    p = re.compile(r"mul\([0-9]+\,[0-9]+\)|don't|do")
    
    must_sum = True
    total = 0
    for m in p.finditer(line2):
        print(f"DEBUG: m = {m}")
        print(m.start(), m.group())
        g = m.group()
        if g.startswith("don't"):
            must_sum = False
        elif g.startswith("do"):
            must_sum = True
        else:
            operands = re.findall("[0-9]+", g)
            print(f"DEBUG: operands={operands}")
            if must_sum:
                total += int(operands[0]) * int(operands[1])
    return total

test_line2 = "xmul(2,4)&mul[3,7]!^don't()_mul(5,5)+mul(32,64](mul(11,8)undo()?mul(8,5))"
print(f"Part 2: TEST: TOTAL = {part2(test_line2)}")
```
