```python
def move_one_step(map:list, m:str, robot_pos:tuple=None) -> bool:
    """
    Return True if cannot move
    """
    # ic("move_one_step", m)
    result = False

    if robot_pos == None:
        robot_pos = find_robot(map)
    rr, rc = robot_pos

    # ic(robot_pos, rr, rc)

    if m == "^":
        ch = map[rr-1][rc]
        if ch == ".":
            # Move robot up
            map[rr-1][rc] = "@"
            map[rr][rc] = "."
        elif ch == "#":
            pass
        elif ch == "O":
            # DEBUG
            n = 1
            while map[rr-n][rc] == "O":
                n += 1
            # ic(rr,rc,n)
            if map[rr-n][rc] == ".":
                # Push train down
                for m in range(n,0,-1):
                    # ic(rr,rc,m,n)
                    ch2 = map[rr-m][rc]
                    map[rr-m][rc] = map[rr-m+1][rc]
                    map[rr-m+1][rc] = ch2
            # END DEBUG
        else:
            assert False, f"TODO: ^ with {ch}"
            ...
    elif m == ">":
        ch = map[rr][rc+1]
        if ch == ".":
            # Move robot right
            map[rr][rc+1] = "@"
            map[rr][rc] = "."
        elif ch == "#":
            pass
        elif ch == "O":
            n = 1
            while map[rr][rc+n] == "O":
                n += 1
            # ic(rr,rc,n)
            if map[rr][rc+n] == ".":
                # Push train right
                for m in range(n,0,-1):
                    # ic(rr,rc,m,n)
                    ch2 = map[rr][rc+m]
                    map[rr][rc+m] = map[rr][rc+m-1]
                    map[rr][rc+m-1] = ch2
        else:
            assert False, f"TODO: > with {ch}"
            ...
    elif m == "v":
        ch = map[rr+1][rc]
        if ch == ".":
            # Move robot down
            map[rr+1][rc] = "@"
            map[rr][rc] = "."
        elif ch == "#":
            pass
        elif ch == "O":
            n = 1
            while map[rr+n][rc] == "O":
                n += 1
            # ic(rr,rc,n)
            if map[rr+n][rc] == ".":
                # Push train down
                for m in range(n,0,-1):
                    # ic(rr,rc,m,n)
                    ch2 = map[rr+m][rc]
                    map[rr+m][rc] = map[rr+m-1][rc]
                    map[rr+m-1][rc] = ch2
        else:
            assert False, f"TODO: v with {ch}"
            ...
    elif m == "<":
        ch = map[rr][rc-1]
        if ch == ".":
            # Move robot left
            map[rr][rc-1] = "@"
            map[rr][rc] = "."
        elif ch == "#":
            pass
        elif ch == "O":
            # DEBUG
            n = 1
            while map[rr][rc-n] == "O":
                n += 1
            # ic(rr,rc,n)
            if map[rr][rc-n] == ".":
                # Push train left
                for m in range(n,0,-1):
                    # ic(rr,rc,m,n)
                    ch2 = map[rr][rc-m]
                    map[rr][rc-m] = map[rr][rc-m+1]
                    map[rr][rc-m+1] = ch2
            # END DEBUG
        else:
            assert False, f"TODO: < with {ch}"
            ...
    else:
        assert False, ic(f"Illegal move {m}")

    return result
```
